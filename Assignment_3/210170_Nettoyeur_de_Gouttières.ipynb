```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
```

```python
def gutter(number):
    image_path = f"gutters{number}.JPG"
    image = cv2.imread(image_path)

    cv2_imshow(image)

    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    height, width = gray.shape[:2]

    print(f"{width} x {height}")

    if number == 1:
        dilated_img = cv2.dilate(gray, np.ones((5, 5)))

        blurred_img = cv2.GaussianBlur(dilated_img, (7, 7), 128)

        final = 255 - cv2.absdiff(blurred_img, gray)

        for i in range(height):
            for j in range(100):
                final[i, j] = (final[i, j] / 255) ** (1.4) * 255

    elif number == 2:
        dilated_img = cv2.dilate(gray, np.ones((19, 19)))

        blurred_img = cv2.medianBlur(dilated_img, 23)

        final = 255 - cv2.absdiff(blurred_img, gray)

    elif number == 3:
        final = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 7, 2)


    # Display the final image
    cv2_imshow(final)
```

```python
# Test the function with different numbers
gutter(1)
gutter(2)
gutter(3)
```
