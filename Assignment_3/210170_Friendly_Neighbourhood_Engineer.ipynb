```python
import cv2 as cv
import numpy as np

```

```python
def enhance_image(cctv_number):

    image_path = f'cctv{cctv_number}.jpg'
    image = cv.imread("/content/cctv3.JPG")

    gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)

    equalized = cv.equalizeHist(gray)

    clahe = cv.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    enhanced = clahe.apply(equalized)

    enhanced_path = f'enhanced-cctv{cctv_number}.jpg'
    cv.imwrite(enhanced_path, enhanced)

    print(f'Enhanced image saved as {enhanced_path}')

cctv_number = int(input('Enter the CCTV number (1, 2, 3, or 4): '))
enhance_image(cctv_number)
```
